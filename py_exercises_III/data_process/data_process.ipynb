```python
data_gender = data['性别']
data_gender.head(3)
```

```python
data_gender_re = data_gender[data_gender.notnull()] 
# data_gender_re.head(4)
data_gender_re.unique()
```

```python
count = len(data_gender_re)
count_m = len(data_gender_re[data_gender_re == '男'])
count_w = len(data_gender_re[data_gender_re == '女'])
rate_m = count_m/count
rate_w = count_w/count
print(' 样本量：' + str(count) + "\n",
      '男性样本量：' + str(count_m) + "\n",
      '女性样本量：' + str(count_w) + "\n",
      '男性比例：' + str(rate_m*100)[:4] + '%' + "\n",
      '女性比例：' + str(rate_w*100)[:4] + '%'
        )
```

```python
data_gender2 = data[['省级政区名称','性别']]
data_gender2_re = data_gender2[data_gender2['性别'].notnull()]
data_gender2_re.head(3)
```

```python
# 按照省份统计出男女人数
# crosstab(行，列)用于针对字符串数据的透视（类似excel的数据透视）
pt = pd.crosstab(data_gender2_re['省级政区名称'],data_gender2_re['性别'])
# pt = pd.crosstab(data_gender2_re['性别'],data_gender2_re['省级政区名称'])
pt.head(5)
```

```python
pt['女性占比'] =  pt['女'] / (pt['女'] + pt['男'])
pt2 = pt.sort_values(by=['女性占比'],ascending=False)
pt2.head(4)
```

```python
matplotlib.matplotlib_fname()
```

```python
fig_1 = plt.figure(figsize=(8,4))
index = pt2.index[:10]
plt.bar(range(10),  # 横坐标
        pt2['女性占比'][:10], # 纵坐标
        tick_label=index, # 横轴标签
        color = 'pink' )  # 颜色
plt.title('不同省份女性市委书记占比')
plt.xlabel('省份')
plt.ylabel('女性占比')
plt.show()
```

```python
fig_2 = plt.figure(figsize=(4,4))

plt.boxplot(pt2['女性占比'], # 值
            vert=True, # true：纵向，false：横向
            showmeans = True) # 显示均值
plt.title('女性市委书记占比结构')
plt.xticks([])
plt.ylabel('女性占比')
plt.show()
```

```python
data_age = data[['出生年份','党委书记姓名','年份']]
data_age_re = data_age[data_age['出生年份'].notnull()]
data_age_re.head(3)
```

```python
df1 = 2017 - data_age_re['出生年份']
df1.head()
```

```python
df_yearmin = data_age_re[['党委书记姓名','年份']].groupby(data_age_re['党委书记姓名']).min() #每个人的入职时候 年份最小值
df2 = df_yearmin['年份'].groupby(df_yearmin['年份']).count() #按照年份分，看每年有多少个
df2
```

```python
df_yearmax = data_age_re[['党委书记姓名','年份']].groupby(data_age_re['党委书记姓名']).max()  #每个人的卸任时候 年份最大值
df3 = df_yearmax['年份'].groupby(df_yearmax['年份']).count()   #按照年份分，看每年有多少个
df3
```
